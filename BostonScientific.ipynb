```python
boston_house=df[['Crime_Rate','Zone','Rooms','Age','Distances','Acessibility','LOW_Stat']]
boston_house.head(5)
```

```python
price=df['MedianValue']
```

```python
complete_dataset=df[['Crime_Rate','Zone','Rooms','Age','Acessibility','LOW_Stat','MedianValue']]
sns.pairplot(complete_dataset)
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
X_train,X_test,Y_train,Y_test=train_test_split(boston_house,price,test_size=0.2,random_state=42)
```

```python
lm=LinearRegression()
```

```python
lm.fit(X_train,Y_train)
```

```python
lm.intercept_
```

```python
lm.coef_
```

```python
cdf=pd.DataFrame(lm.coef_,boston_house.columns,columns=['Coeff'])
cdf.head(10)
```

```python
predictions=lm.predict(X_test)
```

```python
import matplotlib.pyplot as plt
plt.scatter(Y_test,predictions)
```

```python
sns.distplot((Y_test-predictions))
```

```python
from sklearn import metrics
metrics.mean_absolute_error(Y_test,predictions)
```

```python
metrics.mean_squared_error(Y_test,predictions)
```

```python
import numpy as np
np.sqrt(metrics.mean_squared_error(Y_test,predictions))
```

```python
lm.score(X_test,Y_test)
```
